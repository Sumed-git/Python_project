**BAR CHARTS**

```python
job_counts = df.job_title_short.value_counts()
```

```python
job_counts
```

```python
job_counts = job_counts.sort_values(ascending=True)
plt.barh(job_counts.index, job_counts)
plt.show()
```

```python
job_counts = job_counts.sort_values(ascending=False)
plt.bar(job_counts.index, job_counts)
plt.title(' Posting by Jobs by Title')
plt.ylabel('Counts of job Postings')
plt.xticks(rotation=45, ha = 'right')
plt.show()
```

**MATPLOTLIB AND PANDAS**

```python
job_counts.plot(kind='bar')
plt.title(' Posting by Jobs by Title')
plt.ylabel('Counts of job Postings')
plt.xlabel('')
plt.xticks(rotation=45, ha = 'right')
plt.show()
```

**MEDIAN SALARY VS DATA SCIENCE JOB**

```python
df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values().plot(kind='barh')
job_salary.plot(kind='barh')
plt.xlabel('salary ($USD)')
plt.ylabel('')
plt.title('Median Salary by Job Title')
plt.xticks(rotation=45, ha = 'right')
plt.show()
```
