calculate project salary next year

```python
#import libraries
import pandas as pd
from datasets import load_dataset 
import matplotlib.pyplot as plt

# Load Dataset
dataset = load_dataset('Sumedchakma/jobs_record')
df = dataset['train'].to_pandas()

#Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['salary_year_avg']
```

```python
pd.notna(df['salary_year_avg'])
```

```python
df[pd.notna(df['salary_year_avg'])]['salary_year_avg']
```

```python
df_salary = df[pd.notna(df['salary_year_avg'])].copy()

def projected_salary(salary):
    return salary * 1.03

df_salary['updated_salary'] = df_salary['salary_year_avg'].apply(projected_salary)

df_salary[['salary_year_avg', 'updated_salary']]
```
