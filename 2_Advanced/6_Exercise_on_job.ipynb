```python
#import libraries
import pandas as pd
from datasets import load_dataset 
import matplotlib.pyplot as plt

# Load Dataset
dataset = load_dataset('Sumedchakma/jobs_record')
df = dataset['train'].to_pandas()

#Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
df_US
```

```python
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
df_US
```

```python
df_US_pivot = df_US.pivot_table(
    index='job_posted_month',
    columns='job_title_short',
    aggfunc='size'
)
```

```python
df_US_pivot
```

```python
df_US_pivot.reset_index(inplace=True)

df_US_pivot['month_no'] = pd.to_datetime(
    df_US_pivot['job_posted_month'], 
    format='%B'
).dt.month

df_US_pivot.sort_values('month_no', inplace=True)

df_US_pivot.set_index('job_posted_month', inplace=True)

df_US_pivot.drop(columns='month_no', inplace=True)
```

```python
df_US_pivot
```

```python
df_US_pivot.plot(kind='line')
```

```python
top_3 = df_US['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
```

```python
df_US_pivot[top_3].plot(kind='barh')
plt.title('Job posted in US')
plt.xlabel('Job count')
plt.ylabel('In the year 2023')
plt.legend()
plt.show
```

```python
df_US_pivot[top_3].plot(kind='line')
plt.title('Job posted in US')
plt.xlabel('In the year 2023')
plt.ylabel('Job count')
plt.legend()
plt.show
```

```python
df_US_pivot[top_3].sum().plot(kind='pie')
plt.title('Job posted in US (Top 3 Roles)')
plt.ylabel('')
plt.show()
```

```python
df_US_pivot.loc['January', top_3].plot(kind='pie')
plt.title('Job posted in US - January')
plt.ylabel('')
plt.show()
```

```python
df_US_pivot[top_3].sum().plot(
    kind='pie',
    autopct='%1.1f%%'   # shows percentage with 1 decimal
)

plt.title('Job Posted in US (Top 3 Roles - 2023)')
plt.ylabel('')
plt.show()
```
